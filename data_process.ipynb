## Import thư viện

```python
import pandas as pd
import numpy as np
```

## Đọc dữ liệu

```python
df_user = pd.read_excel('data/Survey_data-1.xlsx', sheet_name='data')
df_user.rename(columns={'id': 'user_id'}, inplace=True)
df_user.head()
```

```python
df_aspiration = pd.read_excel('data/Survey_data-1.xlsx', sheet_name='aspiration')
df_aspiration.rename(columns={'id_data': 'user_id'}, inplace=True)
df_aspiration.head()
```

```python
df_user_meger = df_user.merge(df_aspiration, on='user_id', how = 'left')
df_user_meger.drop(columns= ['id'], inplace=True)
df_user_meger.head()
```

```python
df_uni = pd.read_excel('data/Data_Uni.xlsx', sheet_name='UNI')
df_uni.rename(columns={'ID': 'uni_id'}, inplace=True)
df_uni.head()
```

```python
df_user_meger_uni = df_user_meger.merge(df_uni[['uni_id', 'Uni']], how='left', left_on=df_user_meger['university'].str.lower(), right_on=df_uni['Uni'].str.lower())
df_user_meger_uni.drop(columns = ['Uni', 'key_0'], inplace = True)
df_user_meger_uni.head()
```

```python
df_major = pd.read_excel('data/Data_Uni.xlsx', sheet_name='Uni_major')
df_major.drop(columns=['Unnamed: 6','Unnamed: 7'], inplace=True)
df_major['ID_uni'].astype('float')
df_major.head()
```

# Phân tích và xử lý dữ liệu

```python
df_user_meger_uni.info()
```

### Các trường hợp missing là do người dùng chỉ điền "Trường đại học Mở" thay vì "Trường đại học Mở Hà Nội" => 62

```python
df_user_meger_uni['uni_id'].fillna(62, inplace=True)
```

```python
df_user_meger_uni.head()
```

```python
df_user_meger_uni.info()
```

Xóa cột: 'name', 'university'

```python
df_user_meger_uni.drop(columns=['name', 'university'], inplace=True)
df_user_meger_uni.head()
```

```python
df_user_meger_uni['major_id'] = None
for i in range(0, len(df_user_meger_uni)):
    for j in range(0, len(df_major)):
        if int(df_user_meger_uni['uni_id'][i]) == df_major['ID_uni'][j]: #Tim truong dai hoc xem ton tai hay ko
            if df_user_meger_uni['major'][i].lower() == df_major['Major'][j].lower(): #nganh co ton tai hay ko
                df_user_meger_uni['major_id'][i] = df_major['ID_major'][j] #lay ma nganh tuong ung 
```

```python
df_user_meger_uni.info()
```

Điền -1 vào các giá trị major_id bị khuyết thiếu

```python
df_user_meger_uni['major_id'].fillna(-1, inplace=True)
df_user_meger_uni.drop(columns=['major'], inplace=True)
```

```python
df_user_meger_uni.head(6)
```

```python
len(set(list(df_user_meger_uni["major_id"])))
```

```python
df_user_meger_uni['block'] = df_user_meger_uni['block'].str.split(',').str.get(0)
```

Danh sách học lực encode

```python
from sklearn.preprocessing import LabelEncoder

academic_encoder = LabelEncoder()
df_user_meger_uni['academic_encode'] = academic_encoder.fit_transform(df_user_meger_uni['academic'])
academic_encoder.classes_
```

Danh sách tính cách encode

```python
personality_encoder = LabelEncoder()
df_user_meger_uni['personality_encode'] = personality_encoder.fit_transform(df_user_meger_uni['personality'])
personality_encoder.classes_
```

Danh sách ngành encode

```python
major_id_encoder = LabelEncoder()
df_user_meger_uni['major_id_encode'] = major_id_encoder.fit_transform(df_user_meger_uni['major_id'].astype('str'))
major_id_encoder.classes_
```

Danh sách khối encode

```python
block_encoder = LabelEncoder()
df_user_meger_uni['block_encode'] = block_encoder.fit_transform(df_user_meger_uni['block'])
block_encoder.classes_
```

```python
df_user_meger_uni.head()
```

```python
df_user_meger_uni.columns
```

Chuyển dữ liệu về đúng kiểu dữ liệu

```python
df_user_meger_uni.drop(columns=['academic', 'personality', 'block',  'major_id'], inplace=True)
df_user_meger_uni['user_id'] = df_user_meger_uni['user_id'].astype(np.int16)
df_user_meger_uni['isLike'] = df_user_meger_uni['isLike'].astype(np.int16)
df_user_meger_uni.head()
```

```python
df_user_meger_uni.to_csv('data/data_processed.csv', index = False)
```
